```python
import speech_recognition as sr  #speech to text
import pyttsx3 as pt    #text to speech
import datetime
import pandas as pd
from tabulate import tabulate

engine = pt.init("sapi5")
voices = engine.getProperty("voices")
engine.setProperty("voice",voices[1].id)

class Twinkle:
    def speak(self,audio):
        engine.say(audio)
        engine.runAndWait()

    def wishme(self):
        hour = int(datetime.datetime.now().hour)
        if hour>=0 and hour<12:
            self.speak("Good Morning")
        elif hour>=12 and hour<18:
            self.speak("Good Afternoon")
        else:
            self.speak("Good Evening")

        self.speak("I am Twinkle, please tell me how may i help you?")

    def takecommand(self):
        r=sr.Recognizer()
        with sr.Microphone() as source:
            r.adjust_for_ambient_noise(source,duration=1)
            print("Listening..")
            audio=r.listen(source)
        try:
            print("Recognizing..")
            query=r.recognize_google(audio,language="en-in")
            print(f"User Query: {query}")
        except:
            print("say that again please..")
            return "None"
        return query

class pandas(Twinkle): 
    def upload(self,path):
        self.df = pd.read_csv(path)
        print("CSV FILE UPLOAD SUCCESSFULLY..")
        self.speak("CSV FILE UPLOAD SUCCESSFULLY")
        print("__________________________________________")
    
    def columns(self):
        print("Dataset Contain Columns are:")
        self.speak("Dataset Contain Columns are:")
        self.x=[i for i in self.df.columns ]
        for i in self.x:
            print(i)
        print(" __________________________________________")
    
    def null_values(self):
        j=0
        self.speak("check for null values in given dataset")
        for i in  self.df.isnull().sum():
            print(f"{self.x[j].upper()} contain {i} null values..")
            j+=1
        print("__________________________________________")
        
    def shape(self):
        self.speak(f"shape of the dataset is {self.df.shape[0]} rows and {self.df.shape[1]} columns")
        print(f"shape of the dataset is {self.df.shape[0]} rows and {self.df.shape[1]} columns")
        print("__________________________________________")
        
    def head(self):
        self.speak("Display first few Rows in given dataset")
        print(tabulate(self.df.head(5),headers="keys",tablefmt="pretty"))
        print("__________________________________________")

ai = Twinkle()
obj = pandas()   #r"D:\MACHINE LEARNING\Assignmnets\house_prices.csv"

ai.speak("Please upload CSV FILE")
path=input("Enter csv file Path: ")
obj.upload(path)

ai.wishme()

while True:
    query=ai.takecommand().lower()

    if "who are you" in query:
        ai.speak("I am Your Virtual Assistant sir, My name is Twinkle")

    elif "what can you do" in query:
        ai.speak("I can search google, open Instagram, Tell joke, Tell time")

    elif "time now" in query:
        time_now = datetime.datetime.now().strftime("%H:%M:%S")
        ai.speak(f"time now is{time_now}")
    
    elif "tell me columns names" in query:
        obj.columns()
        
    elif "check for null values" in query:
        obj.null_values()

    elif "shape of the dataset" in query:
        obj.shape()

    elif "glance about dataset" in query:
        obj.head()
        
    elif "leave" in query:
        break
```
