```python
import random
class ATM_SIMULATOR():
    def __init__(self,username,password,balance):
        self.username = username
        self.password = password
        self.balance = balance
    
    def account_verify(self):
        username = input("Enter Your Username: ")
        password = input("Enter Your Password: ")
        if (username==self.username and self.password==password):
            user_name = username.upper()
            print(f"Welcome to Our ATM Machine! \nHai {user_name}! ğŸ˜ƒ")
            print("______________________________________________________x______________________________________________________")

            return True
        else:
            print("INVALID USERNAME/PASSWORDğŸ˜± \nPlease Try Again!! ğŸ˜‘")
            print("______________________________________________________x______________________________________________________")

            return False
        
    def generate_otp(self):
        otp = random.randint(100000,999999)
        return otp
        
    def otp_verify(self,otp):
        user_otp = int(input("Enter OTP number: "))
        if (otp==user_otp):
            print("WowğŸ¥³ğŸ‘! OTP verification SuccessfulğŸ¥³ğŸ‘!!")
            print("______________________________________________________x______________________________________________________")

            return True
        else: 
            print("Sorry! OTP verification FaildğŸ‘!!\nPlease Try Again!! ğŸ˜‘")
            print("______________________________________________________x______________________________________________________")

            return False
        
    def check_balance(self):
        print(f"Your Account Current Balance: {self.balance:,}ğŸ¤‘")
        print("______________________________________________________x______________________________________________________")

        
    def deposit(self):
        deposit_amt = int(input("Enter amount to deposit: "))
        self.balance += deposit_amt
        print(f"Deposit Successful!ğŸ¥³ğŸ‘\nCurrent Balance is {self.balance:,}ğŸ¤‘")
        print("______________________________________________________x______________________________________________________")

        
    def withdraw(self):
        withdraw_amt = int(input("Enter amount to withdraw: "))
        if(withdraw_amt < self.balance):
            self.balance -= withdraw_amt
            print(f"Withdrawal SuccessfulğŸ¥³ğŸ‘! \n Current Balance is {self.balance:,}ğŸ¤‘")
            print("______________________________________________________x______________________________________________________")

        else:
            print(f"Insufficient BalanceğŸ˜Œ! \n your current balance is {self.balance:,}ğŸ¤‘ only!!")
            print("______________________________________________________x______________________________________________________")
            
    def start_atm_machine(self):
        if self.account_verify():
            condition=1
            while(condition):
                print("""
                Please Choose an Option:\n
                1. Check Balance
                2. Deposit
                3. Withdraw
                4. Exit
                 """)

                choice = int(input("Enter option number: "))
                
                if choice == 4:
                    print("Thanking you for using our ATM MachineğŸ¥°\nSee You AgainğŸ˜‡")
                    print("______________________________________________________x______________________________________________________")

                    break
                elif choice not in [1,2,3,4]:
                    print("Invalid optionğŸ˜‘")
                    print("______________________________________________________x______________________________________________________")
                    break

                otp = self.generate_otp()
                print(f"OTP sent to your ATM registered number {otp}")
               
                if not self.otp_verify(otp):
                    condition = 0
                else:
                    if choice == 1:
                        self.check_balance()
                    elif choice == 2:
                        self.deposit()
                    elif choice == 3:
                        self.withdraw()

obj=ATM_SIMULATOR("aravindh","123",200000)
```
